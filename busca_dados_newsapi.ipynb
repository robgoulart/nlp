```python
import pandas as pd
import requests
from math import ceil

#variaveis para construção do arquivo de dados
arquivo='/media/dados/Nextcloud/lab/NLP/newsapi_articles.xlsx'
writer=pd.ExcelWriter(arquivo, engine='openpyxl') #necessario instalar pacote openpyxl
df=pd.DataFrame(columns=['Data', 'Titulo', 'Descricao', 'Conteudo'])

#variaveis para API
language='pt'
termo='bolsonaro'
dt_from='2020-08-28'
dt_to='2020-08-28'
sort_by='popularity'
key='543ff7e878d646c189ecf7d3e70ef916'
url='http://newsapi.org/v2/everything?q=%s&language=%s&from=%s&to=%s&sortBy=%s&apiKey=%s&pageSize=100&page=1' % (termo, language, dt_from, dt_to, sort_by, key)

#tratando os resultados
data=requests.get(url)

if data.json()['status']=='ok':
    totalResults=data.json()['totalResults']
    pgs=ceil(totalResults/100)
    for pg in range(1, pgs):
        #artigos daquela pagina
        url='http://newsapi.org/v2/everything?q=%s&language=%s&from=%s&to=%s&sortBy=%s&apiKey=%s&pageSize=100&page=%s' % (termo, language, dt_from, dt_to, sort_by, key, pg)
        data=requests.get(url)
        print(data.json())
        articles=data.json()['articles']
        for article in articles:
            df=df.append({'Data':article['publishedAt'], 
                 'Titulo': article['title'], 
                 'Descricao': article['description'], 
                 'Conteudo':article['content']}, ignore_index=True)
    df.to_excel(writer, encoding='utf-8', index=False, sheet_name='articles')

    #salvando arquivo de dados
    writer.save()
```
